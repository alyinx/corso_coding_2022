# Il ciclo `while`
Il ciclo `while` è un secondo esempio di struttura ciclica. Il blocco di codice all'interno del ciclo è eseguito finché una condizione rimane vera.

Ad esempio:

```python
n = 2
while n <= 1000:
    print(n)
    n = n*n

# Il codice non indentato non fa parte del ciclo e viene eseguito dopo il ciclo
print("Ciclo terminato.")
print(f"Valore di n dopo il ciclo: {n}") # Nota la f iniziale e la formattazione del testo
```

Dopo l'ultima iterazione del ciclo, `n` ha assunto il valore $256^2 = 65536 > 1000$. Quindi, la condizione del ciclo `while`, $n \leq 1000$, è diventata falsa e l'iterazione si è interrotta. In seguito, dopo il ciclo, sono stati eseguiti i due comandi `print`.

## Indovina il numero

Il seguente codice genera un numero segreto e chiede all'utente di indovinarlo:

```python
from random import *

segreto = randint(1, 100)

tentativo = -1

print("Indovina il numero segreto!")

while segreto != tentativo:
    tentativo = int(  input("Digita un numero compreso tra 1 e 100: ")  )
    
    if tentativo > segreto:
        print("Errore! Il numero segreto è minore.")
        
    if tentativo < segreto:
        print("Errore! Il numero segreto è maggiore.")

print("Complimenti, hai indovinato il numero!")
```

## Il moto browniano
Il seguente codice fa muovere casualmente la tartaruga. Un moto di questo tipo è detto browniano.

Il moto della tartaruga è terminato quanto la tartaruga esce da un quadrato di lato 400 pixel centrato nell'origine.

```python
from turtle import *
from math import *
from random import *

shape("turtle")
home()
clear()
speed(10)

# coordinate
x = xcor()
y = ycor()

while abs(x) < 200 and abs(y) < 200:
    title(f"Coordinate x = {x:.1f} y = {y:.1f}") # titolo della finestra con coordinate
    l = randint(0, 20)
    alpha = randint(0, 359)
    
    #movimento della tartaruga
    setheading(alpha)
    forward(l)
    
    x = xcor()
    y = ycor()
```

Nota alcune novità:
* la funzione `abs` calcola il valore assoluto;
* `and` è un connettivo logico, serve per calcolare l'intersezione logica di due proposizioni. `or` è il connettivo logico che serve per calcolare l'unione logica.
* nota il titolo della finestra, che è una **stringa formattata** (che inizia con la lettera `f" "`) e contiene i valori di due variabili.
